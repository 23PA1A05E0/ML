#4. Histogram


o Simulate 100 random values from a normal distribution using np.random.randn() and plot a histogram.



```python
# Simulate 100 random values from a normal distribution
data = np.random.randn(100)
plt.hist(data, edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of 100 Random Values from a Normal Distribution')

# Display the plot
plt.show()
```

o Customize the number of bins and add a grid to the plot.


```python
# Simulate 100 random values from a normal distribution
data = np.random.randn(100)
plt.hist(data, bins=15, edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of 100 Random Values from a Normal Distribution')
plt.grid(True)
plt.show()
```

#6. Box Plot (Seaborn)

o Create a box plot to visualize the distribution of the marks scored by
students in a class: [78, 85, 92, 88, 76, 95, 89, 84, 91, 73].

```python
import seaborn as sns
marks = [78, 85, 92, 88, 76, 95, 89, 84, 91, 73]
sns.boxplot(data=marks)
plt.xlabel('Students')
plt.ylabel('Marks')
plt.title('Box Plot of Marks Scored by Students')
# Display the plot
plt.show()
```

o Highlight the median and any outliers.


```python
marks = [78, 85, 92, 88, 76, 95, 89, 84, 91, 73]

sns.boxplot(data=marks)

median = np.median(marks)
plt.axhline(median, color='r', linestyle='--', label='Median')

plt.xlabel('Students')
plt.ylabel('Marks')
plt.title('Box Plot of Marks Scored by Students')

plt.legend()
plt.show()
```

#8. Comparison of Matplotlib and Seaborn

o Generate a pair of plots (bar charts or line plots) using the same data in
both Matplotlib and Seaborn.

o Highlight the differences in the code, syntax, and aesthetics of the two
libraries.

```python
import matplotlib.pyplot as plt
categories = ['A', 'B', 'C', 'D', 'E']
values = [20, 30, 25, 15, 10]
plt.bar(categories, values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart using Matplotlib')
# Display the plot
plt.show()
```

#Differences in Code, Syntax, and Aesthetics

#Code and Syntax:
Matplotlib: The code is more straightforward and requires fewer imports. You directly use the plt.bar() function to create the bar chart.

**Seaborn**: You need to create a DataFrame to use Seaborn effectively. The sns.barplot() function is used, and you pass the DataFrame along with the column names for the x and y axes.

#Aesthetics:
***Matplotlib***: The default aesthetics are simpler and more basic. Customization requires more manual adjustments.

***Seaborn***: The default aesthetics are more polished and visually appealing. Seaborn automatically applies a more sophisticated style to the plots.

#10. Save and Style Plots

‚Ä¢ Plot any graph of your choice and save it in both .png and .pdf formats.

‚Ä¢ Apply at least one built-in style (e.g., 'ggplot' or 'seaborn-darkgrid') and explain
how the style impacts the visualization.

```python
import matplotlib.pyplot as plt
import numpy as np

# Apply the 'ggplot' style
plt.style.use('ggplot')

# Generate x values
x = np.linspace(0, 10, 100)

# Compute y values
y = np.sin(x)

# Create the plot
plt.plot(x, y, label='sin(x)')

# Customize the plot
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.title('Plot of sin(x)')
plt.legend()

# Save the plot in .png and .pdf formats
plt.savefig('plot.png')
plt.savefig('plot.pdf')

# Display the plot
plt.show()
```



### The **'ggplot'** style is inspired by the aesthetics of the ggplot2 package in R.

It provides a clean and modern look to the plots with grid lines, a light gray background, polished font and line styles, and a vibrant color palette.
