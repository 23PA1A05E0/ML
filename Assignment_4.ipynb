#TOP INDIAN PLACES TO VISIT

First importing necessary packages

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Storing the information in data DataFrame



```python
data = pd.read_csv('/content/Top Indian Places to Visit.csv')
```

Checking the no of null values in the given dataset of each column.(To know where the given data is good or bad)

```python
data.isnull().sum()
```

we are dealing some sample data from given dataset

```python
sample = data.sample(10)
sample
```

to get basic stats of taken dataset

```python
sample.describe()
```

```python
sample.info()
```

To know How many hours it take to cover all things in a particular place
**VAISHNO DEVI** is taking high amount of time

```python
plt.figure(figsize=(10,5))
x = sample['Name']
y = sample['time needed to visit in hrs']
plt.bar(x,y,color="lightblue")
plt.xticks(rotation=90)
plt.xlabel("Name")
plt.ylabel("time needed to visit in hrs")
plt.legend()
plt.title("Time needed to visit in hrs")
plt.show()
```

Here the below graph tells **Best place to visit** based on people rating

```python
x = sample['Name']
y = sample['Number of google review in lakhs']
plt.figure(figsize=(10,5))
plt.subplot(1,1,1)
plt.bar(x,y,color='gold')
plt.xlabel("Name")
plt.xticks(rotation=90)
plt.ylabel("Rating")
plt.legend()
plt.title("Best places to go")
plt.show()
```

This gives us details gives which particular time to go to visit the place for best exeprience .

```python
y = sample['Best Time to visit']
x = sample['Name']
plt.figure(figsize=(10,5))
plt.bar(x,y,color='#456')
plt.xlabel("Best Time to visit")
plt.xticks(rotation=90)
plt.show()
```

Here weather the **CAMERA** is allowed or not

```python
import matplotlib.pyplot as plt

x = sample['Name']
y = sample['DSLR Allowed']

plt.figure(figsize=(8, 5))
sns.scatterplot(x=x, y=y,data=sample)
plt.xlabel("Name")
plt.ylabel("DSLR Allowed")
plt.title("DSLR Allowed by Name")
plt.xticks(rotation=90)  # Rotate x-axis labels if necessary
plt.show()
```

Here it tells us about what the place is means temple,Nature etc

```python
x = sample['Name']
y = sample['Significance']
plt.figure(figsize=(10,5))
plt.bar(x,y,color='lightgreen')
plt.xlabel("Name")
plt.xticks(rotation=90)
plt.ylabel("Significance")
plt.ylim(bottom=0)

plt.title("Type of places")
plt.show()
```

The flexibility of travelling tells in below.

```python
x = sample['Name']
y = sample['Airport with 50km Radius']
plt.figure(figsize=(10,5))
plt.subplot(1,1,1)
sns.scatterplot(x=x,y=y,data=sample)
plt.xlabel("Name")
plt.xticks(rotation=90)
plt.ylabel("Airport with 50km Radius")
plt.title("Airport with 50km Radius")
plt.show()
```

Entry fee to visit the tourist place

```python
x = sample['Name']
y = sample['Entrance Fee in INR']
plt.figure(figsize=(10,5))
plt.bar(x,y)
plt.xlabel("Name")
plt.xticks(rotation=90)
plt.ylabel("Entrance Fee in INR")
plt.title("Entrance Fee in INR")
plt.show()
```

The places belongs to which state.

```python
x = sample['Name']
y = sample['State']
sns.scatterplot(x=x,y=y,data=sample)
plt.xlabel("Name")
plt.xticks(rotation=90)
plt.ylabel("state")
plt.title("state")
plt.show()
```

From this,I knew that many of place have zero entry fee .Every state have atleast one tourist place.It's also a income source to the state.Many tourist places are near to aiports which creates easy transportation.
